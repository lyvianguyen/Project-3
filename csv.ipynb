```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

```python
# Convert 'Deaths' column to integers
df['Deaths'] = df['Deaths'].astype(int)

# Filter data for United States
US_data = df[df['State'] == 'United States']

# Group data by 'Cause Name' and sum the number of deaths
cause_data = US_data.groupby('Cause Name')['Deaths'].sum()

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(cause_data, labels=cause_data.index, autopct='%1.1f%%', startangle=140)
plt.title('Causes of Death in the United States')
plt.axis('equal') 

plt.savefig("Causes of Death in the United States")
plt.show()
```

```python
df['Deaths'] = df['Deaths'].astype(int)
grouped_data = df.groupby(['State', 'Cause Name'])['Deaths'].sum().reset_index()
grouped_data = grouped_data[grouped_data['State']!="United States"]
sorted_data = grouped_data.sort_values(by='Deaths',ascending=False)
sorted_data
```

```python
# Group data by state and cause name, and sum the number of deaths
#df['Deaths'] = df['Deaths'].astype(int)
#grouped_data = df.groupby(['State', 'Cause Name'])['Deaths'].sum().reset_index()
#grouped_data = grouped_data[grouped_data['State']!="United States"]
# Pivot the data for better visualization
pivot_data = sorted_data.pivot(index='State', columns='Cause Name', values='Deaths').fillna(0)
# pivot_data.drop(['All causes'], axis=1, inplace=True)
# Creating the bar chart for Deaths per State and Cause Name
pivot_data.plot(kind='bar', stacked=True, figsize=(15, 8))
plt.title('Cause of Deaths per State')
plt.xlabel('State')
plt.ylabel('Number of Deaths')

plt.savefig("Causes of Death per State")
plt.show()
```

```python
# Group data by state and 113 cause name, and sum the number of deaths
df['Deaths'] = df['Deaths'].astype(int)
grouped_data = df.groupby(['State', '113 Cause Name'])['Deaths'].sum().reset_index()
grouped_data = grouped_data[grouped_data['State']!="United States"]
#grouped_data = grouped_data.sort_values(by = ['Deaths'],ascending = False)

# Pivot the data for better visualization
pivot_data = grouped_data.pivot(index='State', columns='113 Cause Name', values='Deaths').fillna(0)

# Creating the bar chart for Deaths per State and 113 Cause Name
pivot_data.plot(kind='bar', stacked=True, figsize=(15, 8))
plt.title('Deaths per State and Cause Name')
plt.xlabel('State')
plt.ylabel('Number of Deaths')
plt.show()
```

```python
# Convert 'deaths' column to integers
df['Deaths'] = df['Deaths'].astype(int)

# Group data by year and cause name, and sum the number of deaths
grouped_data = df.groupby(['Year', 'Cause Name'])['Deaths'].sum().reset_index()

# Create a line chart for trend over time
plt.figure(figsize=(12, 6))
for cause, group in grouped_data.groupby('Cause Name'):
    plt.plot(group['Year'], group['Deaths'], label=cause)

plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.title('Causes of Death Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig("Causes of Death Over Time")
plt.show()
```

```python
# #total deaths over the years
grouped_data = df.groupby(['Year'])['Deaths'].sum().reset_index()

# Create a line chart for trend over time
plt.figure(figsize=(12, 6))

plt.plot(grouped_data['Year'], grouped_data['Deaths'], label='Total Deaths', color='blue')

plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.title('Total Deaths Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig("Total Deaths Over Time")
plt.show()
```

```python
# Group data by state and cause name, and sum the number of deaths
grouped_data = df.groupby(['State', 'Cause Name'])['Deaths'].sum().reset_index()
grouped_data = grouped_data[grouped_data['State']!="United States"]

# Create a bar chart
fig, ax = plt.subplots(figsize=(15, 8))

# Iterate through each state and create a bar for each cause name
for state, group in grouped_data.groupby('State'):
    causes = group['Cause Name']
    deaths = group['Deaths']
    ax.bar(causes, deaths, label=state)

# Set labels and title
ax.set_xlabel('Cause of Death')
ax.set_ylabel('Number of Deaths')
ax.set_title('Cause of Death per State')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")

# Add legend
ax.legend(title='State')

# Show the chart
plt.show()
```

```python
#Creating the scatter plot for Deaths per State and Cause Name
for cause_name, group in grouped_data.groupby('Cause Name'):
    plt.scatter(group['State'], group['Deaths'], label= cause_name)

# Add labels and title
plt.xlabel('State')
plt.ylabel('Number of Deaths')
plt.title('Cause of Death per State')
plt.xticks(rotation=90)
plt.rcParams['figure.figsize']= (18,12)
plt.legend()

# Show the plot
plt.savefig("Cause of Death per State")
plt.show()
```

```python
import pandas as pd
import requests

# Fetch the JSON data from the CDC URL
url = "https://data.cdc.gov/api/views/bi63-dtpu/rows.json?accessType=DOWNLOAD"
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data['data'])
else:
    print("Failed to fetch JSON data from the URL.")

# Specify the column names (found in the 'meta' section of the JSON data)
column_names = [column['name'] for column in data['meta']['view']['columns']]
df.columns = column_names

# Save DataFrame to a CSV file
df.to_csv('cdc_data.csv', index=False)
```
